```python
import os
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import GridSearchCV
```

```python
data=pd.read_excel("Data.xlsx", sheet_name='all')#Select regular or time-saving
data
```

```python
target= data.columns[33:35]
target
```

```python
param_grid = {
    'n_components': [2, 3, 4],
    'covariance_type': ['full', 'tied', 'diag', 'spherical'],
    'n_init': [4,5,6],
    'reg_covar': [0.1, 1.0,10],
}
```

```python
random_state = 112

gmm = GaussianMixture(random_state=random_state)
```

```python
grid_search = GridSearchCV(
    estimator=gmm,
    param_grid=param_grid,
    cv=5,
    verbose=2,
    n_jobs=-1
)
```

```python
# Fit the grid search to the data
grid_search.fit(data[target])
```

```python
# Extract the best parameters and fit a GMM model with a fixed random state
best_params = grid_search.best_params_
gmm_best = GaussianMixture(
    n_components=best_params['n_components'],
    covariance_type=best_params['covariance_type'],
    random_state=random_state
)
```

```python
gmm_best.fit(data[target])
```

```python
result = gmm_best.predict(data[target])
```

```python
result = pd.DataFrame(result)
```

```python
data["Cluster"] = result
data
```
