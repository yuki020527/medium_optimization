```python
data=pd.read_excel("Data.xlsx", sheet_name='regular')#Select regular or time-saving
data
```

```python
features=data.columns[2:33]
features
```

```python
param_grid_gbdt = {"learning_rate":[i/100 for i in range(1,51,1)],  
                   "max_depth":[2,3,4,5],
                   "n_estimators":[300]}

gbdt=GridSearchCV(GradientBoostingRegressor(),param_grid_gbdt,cv=5)
```

```python
gbdt.fit(data[features],data["cells_ml_at_168h"])
                            #"cells_ml_at_96h" was used in time-saving mode 
```

```python
artificial_data=list(itertools.product(c1,c2,c3,c4,c5,c6,c7,c8,c9,c10,c11,c12,c13,c22,c23,c24,c25,c26,c27,c28,c29,c30,c31,c32,c33,c34,c35,c36,c37,c38,c39))
artificial_data=pd.DataFrame(artificial_data,columns=features)
artificial_data
```

```python
predicted_data=gbdt.predict(artificial_data)
```

```python
predicted_reshape = predicted_data.reshape(-1,1)
```

```python
artificial_data_np = np.array(artificial_data)
```

```python
artificial_predicted_data = np.insert(artificial_data_np,[31], predicted_reshape, axis=1)
```

```python
rank_data = artificial_predicted_data[np.argsort(artificial_predicted_data[:,31])][::-1]
```

```python
max_data = rank_data.max()
high_predicted_data = rank_data[np.sum(rank_data,axis=1)>max_data]
```

```python
predicted_result=data.columns[2:34]
high_predicted_result_data = pd.DataFrame(high_predicted_data, columns=predicted_result)
```
